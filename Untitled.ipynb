```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
i=0
while i<2:
    try:
        print(i)
        data = pd.read_excel('PamAir-Data-10-08-2021.xlsx',sheet_name = i)
        i+=1
        list_data=data['AQI'].tolist()
        list_date=data['TIME'].tolist()
        list_data.reverse()
        list_date.reverse()
        #print(data['Location'][0])
        plt.plot(list_date,list_data)
        plt.title(data['Name'][0]+" - "+data['Location'][0])
        plt.xlabel("TimeStamp"+"\n"+"09-08-2021 17:00:00 - 10-08-2021 16:00:00")
        plt.ylabel("AQI")
        namefile='./img_pamair_10_08/'+str(i)+'.jpg'
        plt.savefig(namefile,dpi=1200)
        plt.show()
        print("-----------------------------------------------")
    except:  
        break
```
